```python
df_reg = pd.read_csv("Data/survey_results_public.csv")
df_reg = df_reg[df_reg.Country == 'United States']
df_reg = df_reg[['Age','ConvertedComp','Gender','Ethnicity', 'YearsCodePro', 'Sexuality']]
df_reg["YearsCodePro"].replace({"Less than 1 year": 0, "More than 50 years":50}, inplace=True)
df_reg["YearsCodePro"] = df_reg["YearsCodePro"].apply(pd.to_numeric)
df_reg = df_reg.dropna()
```

```python
df_reg.loc[~df_reg.Gender.isin(['Man','Woman']), 'Gender'] = 'Non-binary'
df_reg.loc[~df_reg.Sexuality.isin(['Straight / Heterosexual']), 'Sexuality'] = 'Other'
df_reg.loc[~df_reg.Ethnicity.isin(['White or of European descent','South Asian','East Asian','Black or of African descent', 'Hispanic or Latino/a/x']), 'Ethnicity'] = 'Other'
```

```python
plt.figure(figsize=(26, 6))
df_reg.groupby(['Ethnicity'])['Ethnicity'].count().plot.bar()
```

```python
len(df_reg)
```

```python
Gender_encode = pd.get_dummies(df_reg['Gender'])
Sexuality_encode = pd.get_dummies(df_reg['Sexuality'], prefix='Sexuality')
Ethnicity_encode = pd.get_dummies(df_reg['Ethnicity'], prefix='Ethnicity')  
```

```python
df_reg = df_reg.join(Gender_encode)
df_reg = df_reg.join(Sexuality_encode)
df_reg = df_reg.join(Ethnicity_encode)
```

```python
df_reg.head()
```

```python
df_reg = df_reg[['ConvertedComp','Age','YearsCodePro','Woman','Non-binary','Sexuality_Other','Ethnicity_Other','Ethnicity_Black or of African descent','Ethnicity_East Asian','Ethnicity_Hispanic or Latino/a/x','Ethnicity_South Asian']]
```

```python
df_reg.head()
```

```python
df_reg = df_reg.rename(columns={"Non-binary": "Non_binary", "Ethnicity_Black or of African descent":"Ethnicity_Black_or_African","Ethnicity_East Asian":"Ethnicity_East_Asian", "Ethnicity_South Asian":"Ethnicity_South_Asian", "Ethnicity_Hispanic or Latino/a/x":"Ethnicity_Hispanic_or_Latino"})
```

```python
df_reg = df_reg.dropna()
```

```python
Q1 = df_reg['ConvertedComp'].quantile(0.05)
Q3 = df_reg['ConvertedComp'].quantile(0.95)
df_reg = df_reg[~((df_reg['ConvertedComp'] < Q1) |(df_reg['ConvertedComp'] > Q3))]
```
