#### Does money really bring job satisfaction ?

```python
df_sat = pd.read_csv("Data/survey_results_public.csv")
df_sat = df_sat[df_sat.Country == 'United States']
Q1 = df_sat['ConvertedComp'].quantile(0.05)
Q3 = df_sat['ConvertedComp'].quantile(0.95)
df_sat = df_sat[~((df_sat['ConvertedComp'] < Q1) |(df_sat['ConvertedComp'] > Q3))]
```

Plotting the salaries of the respondents accoring to their response on job satisfaction

```python
sns.catplot(x="JobSat", y="ConvertedComp", kind="bar", data=df_sat, height=8.27, aspect=11.7/8.27, order=['Very dissatisfied', 'Slightly dissatisfied', 'Neither satisfied nor dissatisfied', 'Slightly satisfied', 'Very satisfied'])
```

**Preparation to do the t-test**

The following section would prepare data for t-test and print the results

```python
s1 = np.array(df_sat[df_sat.JobSat=='Very dissatisfied'].ConvertedComp)
s2 = np.array(df_sat[df_sat.JobSat=='Neither satisfied nor dissatisfied'].ConvertedComp)
s3 = np.array(df_sat[df_sat.JobSat=='Very satisfied'].ConvertedComp)

s1 = s1[~np.isnan(s1)]
s2 = s2[~np.isnan(s2)]
s3 = s3[~np.isnan(s3)]
```

```python
print(np.mean(s1))
print(np.mean(s2))
print(np.mean(s3))

from scipy.stats import ttest_ind
import numpy as np

ttest,pval = ttest_ind(s1,s2)
print("p-value",pval)

ttest,pval = ttest_ind(s3,s2)
print("p-value",pval)

ttest,pval = ttest_ind(s1,s3)
print("p-value",pval)
```

#### How does salary equate to your demographics?

As explored in the report above, this section would cover the linear regression and try to fit the model over the data to come with an equation for the compensation

```python
df_reg = pd.read_csv("Data/survey_results_public.csv")
df_reg = df_reg[df_reg.Country == 'United States']
df_reg = df_reg[['Age','ConvertedComp','Gender','Ethnicity', 'YearsCodePro', 'Sexuality']]
df_reg["YearsCodePro"].replace({"Less than 1 year": 0, "More than 50 years":50}, inplace=True)
df_reg["YearsCodePro"] = df_reg["YearsCodePro"].apply(pd.to_numeric)
df_reg = df_reg.dropna()
df_reg.loc[~df_reg.Gender.isin(['Man','Woman']), 'Gender'] = 'Non-binary'
df_reg.loc[~df_reg.Sexuality.isin(['Straight / Heterosexual']), 'Sexuality'] = 'Other'
df_reg.loc[~df_reg.Ethnicity.isin(['White or of European descent','South Asian','East Asian','Black or of African descent', 'Hispanic or Latino/a/x']), 'Ethnicity'] = 'Other'
Gender_encode = pd.get_dummies(df_reg['Gender'])
Sexuality_encode = pd.get_dummies(df_reg['Sexuality'], prefix='Sexuality')
Ethnicity_encode = pd.get_dummies(df_reg['Ethnicity'], prefix='Ethnicity')  
df_reg = df_reg.join(Gender_encode)
df_reg = df_reg.join(Sexuality_encode)
df_reg = df_reg.join(Ethnicity_encode)
df_reg = df_reg[['ConvertedComp','Age','YearsCodePro','Woman','Non-binary','Sexuality_Other','Ethnicity_Other','Ethnicity_Black or of African descent','Ethnicity_East Asian','Ethnicity_Hispanic or Latino/a/x','Ethnicity_South Asian']]
df_reg = df_reg.rename(columns={"Non-binary": "Non_binary", "Ethnicity_Black or of African descent":"Ethnicity_Black_or_African","Ethnicity_East Asian":"Ethnicity_East_Asian", "Ethnicity_South Asian":"Ethnicity_South_Asian", "Ethnicity_Hispanic or Latino/a/x":"Ethnicity_Hispanic_or_Latino"})
df_reg = df_reg.dropna()
```

**Outlier Treatment**

Taking out the top 5 and bottom 5 percentile from the data to remove the outliers

```python
Q1 = df_reg['ConvertedComp'].quantile(0.05)
Q3 = df_reg['ConvertedComp'].quantile(0.95)
df_reg = df_reg[~((df_reg['ConvertedComp'] < Q1) |(df_reg['ConvertedComp'] > Q3))]
```

**Building the model and printing the results**

```python
import statsmodels.formula.api as sm
result = sm.ols(formula="ConvertedComp ~ YearsCodePro + Woman + Non_binary + Sexuality_Other + Ethnicity_Black_or_African + Ethnicity_Hispanic_or_Latino + Ethnicity_South_Asian + Ethnicity_East_Asian + Ethnicity_Other", data=df_reg).fit()
print(result.params)
```

```python
print(result.summary())
```

**Pay gap across the globe**

In this section we are epxloring if the normalized gender gap is same across the globe or is there any significnat pattern. We can see that it a big issue through out the globe.

```python
df = pd.read_csv("Data/survey_results_public.csv")
df = df.loc[df.Gender.isin(['Man','Woman'])]
df_diff = df.groupby(['Country', 'Gender'])['ConvertedComp'].mean()
df_diff = df_diff.to_frame()
df_diff.reset_index(inplace=True)
temp = df_diff.pivot(index='Country', columns='Gender', values='ConvertedComp')
temp['Woman_perc'] = temp.Woman/temp.Man
temp.reset_index(inplace=True)

import geopandas as gpd
world = gpd.read_file(gpd.datasets.get_path('naturalearth_lowres'))
for_plotting = world.merge(temp, left_on = 'name', right_on = 'Country')
for_plotting = for_plotting.loc[~for_plotting.Country.isin(['Nepal'])]
for_plotting.plot(column='Woman_perc', cmap =    
                                'YlGnBu', figsize=(15,9),   
                                  legend =  
                                  True);
```

**Historical analysis of the gender pay gap**

In this section we will cover the historical analysis of the gender pay gap and draw conclusions based on it.

```python
df15 = pd.read_csv("Data/2015.csv")
df16 = pd.read_csv("Data/2016.csv")
df17 = pd.read_csv("Data/2017.csv")
df18 = pd.read_csv("Data/2018.csv")
df19 = pd.read_csv("Data/2019.csv")
df20 = pd.read_csv("Data/2020.csv")

df15 = df15[['Country','Gender','Compensation: midpoint']]
df16 = df16[['country','gender','salary_midpoint']]
df17 = df17[['Country','Gender','Salary']]
#df18 = df18[['Country','Gender',]]
df19 = df19[['Country','Gender','ConvertedComp']]
df20 = df20[['Country','Gender','ConvertedComp']]

df15 = df15[df15.Country == 'United States']
df16 = df16[df16.country == 'United States']
df17 = df17[df17.Country == 'United States']
df19 = df19[df19.Country == 'United States']
df20 = df20[df20.Country == 'United States']

df15 = df15.rename(columns={'Compensation: midpoint': 'ConvertedComp'})
df16 = df16.rename(columns={'country': 'Country', 'gender': 'Gender', 'salary_midpoint': 'ConvertedComp'})
df17 = df17.rename(columns={'Salary': 'ConvertedComp'})

df15['Year'] = '2015'
df16['Year'] = '2016'
df17['Year'] = '2017'
df19['Year'] = '2019'
df20['Year'] = '2020'

df_hist = df15.append([df16, df17, df19, df20])
df_hist = df_hist.loc[~df_hist.ConvertedComp.isin(['Rather not say','Unemployed'])]
df_hist["ConvertedComp"] = df_hist["ConvertedComp"].apply(pd.to_numeric)

Q1 = df_hist['ConvertedComp'].quantile(0.05)
Q3 = df_hist['ConvertedComp'].quantile(0.95)
df_hist = df_hist[~((df_hist['ConvertedComp'] < Q1) |(df_hist['ConvertedComp'] > Q3))]

df_hist = df_hist.loc[df_hist.Gender.isin(['Man','Woman', 'Male', 'Female'])]
df_hist.loc[df_hist.Gender.isin(['Man','Male']), 'Gender'] = 'Man'
df_hist.loc[df_hist.Gender.isin(['woman','Female']), 'Gender'] = 'Woman'

df_hist["ConvertedComp"] = df_hist["ConvertedComp"].apply(pd.to_numeric)
df_hist = df_hist.groupby(['Year','Gender']).agg({'ConvertedComp':'mean'})

df_hist.reset_index(inplace=True)
df_hist = df_hist.pivot(index='Year', columns='Gender', values='ConvertedComp')
df_hist['Pay_gap'] = ((df_hist.Man - df_hist.Woman)/df_hist.Man)*100
df_hist.reset_index(inplace=True)

plt.plot(df_hist.Year, df_hist.Pay_gap)
plt.ylim(0, 15)
```
